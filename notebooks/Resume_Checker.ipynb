---
## 🔧 Part 1: Environment Setup

Install required packages and configure API credentials.

```python
# ============================================================================
# OPTIONAL: Install Required Packages
# ============================================================================
# Uncomment and run if packages are not installed
# !pip install -qq langchain_groq langchain_community pypdf langchain_core python-dotenv
```

```python
# ============================================================================
# ENVIRONMENT SETUP: Load API Keys
# ============================================================================
import os
from dotenv import load_dotenv

load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")

print("✅ Environment variables loaded successfully!")
```

---
## 🤖 Part 2: LLM Initialization

Initialize the Groq-hosted LLM for resume evaluation.

```python
# ============================================================================
# LLM INITIALIZATION: Configure ChatGroq
# ============================================================================
from langchain_groq import ChatGroq

chat = ChatGroq(
    model="llama-3.3-70b-versatile",  # Fast, capable open-source model
    api_key=GROQ_API_KEY,
)

print("✅ ChatGroq initialized with llama-3.3-70b-versatile")
```

---
## 📄 Part 3: Resume Loading

Load the candidate's PDF resume and extract its text content.

```python
# ============================================================================
# RESUME LOADING: Extract Text from PDF
# ============================================================================
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("./content/Resume.pdf")
documents = loader.load()

context = "\n\n".join(doc.page_content for doc in documents)

print(f"📄 Loaded {len(documents)} page(s) from resume")
```

---
## 📊 Part 4: Evaluation Prompt

Define a structured evaluation prompt that scores the resume across multiple criteria.

### Key Concepts:
- **PromptTemplate** allows reusable prompts with variable injection
- **Multi-criteria scoring** evaluates clarity, relevance, format, comprehensiveness, and keywords
- **Actionable output** includes strengths, weaknesses, and career path suggestions

```python
# ============================================================================
# EVALUATION PROMPT: Multi-Criteria Resume Analysis Template
# ============================================================================
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
    You are an advanced resume evaluation assistant. Analyze the provided
    resume in the context document and score it out of 100 based on the
    following criteria: clarity, relevance, format, comprehensiveness,
    and keywords.

    Your response should be structured as follows:

    1. **Score**: Provide the score out of 100.
    2. **Strengths**: List at least three strengths found in the resume.
    3. **Weaknesses**: List at least three weaknesses or areas for
       improvement in the resume.
    4. **Skills Mentioned**: Identify and list the skills that are
       explicitly mentioned in the resume.
    5. **Recommended Skills**: Suggest additional skills that could
       enhance the resume's effectiveness.
    6. **Next Career Path**: Suggest what should be the next career
       paths for this candidate.

    Please ensure your analysis is clear and concise, providing
    actionable insights for improvement.
    """
)

print("✅ Evaluation prompt template configured")
```

---
## 🚀 Part 5: Run Resume Evaluation

Format the prompt with the resume context and stream the evaluation results.

```python
# ============================================================================
# EVALUATE: Format Prompt and Stream Results
# ============================================================================
question = "Please evaluate this resume."
formatted_prompt = prompt.format(context=context, question=question)

print("🔍 Evaluating resume...\n")

for m in chat.stream(formatted_prompt):
    print(m.content, end="", flush=True)

print("\n\n✅ Resume evaluation complete!")
```
