```python
# ============================================================================
# ENVIRONMENT SETUP: Load API Keys
# ============================================================================
import os
from dotenv import load_dotenv

load_dotenv()

GROQ_API_KEY = os.getenv("GROQ_API_KEY")

print("✅ Environment variables loaded successfully!")
```

---
## 🤖 Part 2: LLM Initialization

Initialize the Groq-hosted LLM that will power our career coach.

```python
# ============================================================================
# LLM INITIALIZATION: Configure ChatGroq
# ============================================================================
from langchain_groq import ChatGroq

chat = ChatGroq(
    model="llama-3.3-70b-versatile",  # Fast, capable open-source model
    api_key=GROQ_API_KEY,
)

print("✅ ChatGroq initialized with llama-3.3-70b-versatile")
```

---
## 📄 Part 3: Resume Loading

Load the candidate's PDF resume and extract its text content for use as context.

```python
# ============================================================================
# RESUME LOADING: Extract Text from PDF
# ============================================================================
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("./content/Resume.pdf")
documents = loader.load()

context = "\n\n".join(doc.page_content for doc in documents)

print(f"📄 Loaded {len(documents)} page(s) from resume")
```

---
## 💬 Part 4: Career Coach System Prompt

Define the system message that shapes the LLM's persona as a professional career coach.
The resume content is injected directly into the system prompt so the coach has full context.

### Key Concepts:
- **SystemMessage** - Sets the AI's behavior and persona
- **HumanMessage / AIMessage** - Track the conversation turns
- **Chat History** - Maintains context across multiple exchanges

```python
# ============================================================================
# SYSTEM PROMPT: Define Career Coach Persona
# ============================================================================
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

chat_history = []

system_message = SystemMessage(
    content=f"""
    You are a professional career coach and resume mentor.

    You help with:
    - Career Guidance
    - Resume Improvements
    - Interview Preparation
    - Job Search Strategy
    - Skill Gap Analysis

    Candidate Resume:
    {context}
    """
)

print("✅ Career coach system prompt configured")
```

---
## 🚀 Part 5: Interactive Chat Loop

Run an interactive conversation with the AI career coach.
Type your questions and get streaming responses. Type `exit` or `quit` to end the session.

```python
# ============================================================================
# INTERACTIVE CHAT: Streaming Conversation Loop
# ============================================================================
while True:
    user_input = input("You: ")

    if user_input.lower() in ["exit", "quit"]:
        print("\n👋 Goodbye!!")
        break

    # Add user message to memory
    chat_history.append(HumanMessage(content=user_input))

    # Build message list
    messages = [system_message] + chat_history

    print("\n🤖 Coach:", end=" ", flush=True)

    response_text = ""

    for chunk in chat.stream(messages):
        print(chunk.content, end="", flush=True)
        response_text += chunk.content

    # Add AI response to memory
    chat_history.append(AIMessage(content=response_text))

    print("\n")
```
